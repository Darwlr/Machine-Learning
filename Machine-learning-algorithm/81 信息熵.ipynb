## 信息熵

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 信息熵
def entropy(p):
    return -p * np.log(p) - (1-p) * np.log(1-p)
```

```python
x = np.linspace(0.01, 0.99, 200) # x不为0和1，log0为无穷
```

```python
plt.plot(x, entropy(x))
```

该图像是以0.5为对称轴的，意味着当 x=0.5 时，这根曲线得到最大值。如果对于信息熵只有两个类别，当一个类别时1/2，另一个也是1/2时，此时整个数据的信息熵就是最大，此时数据是最不稳定的，确定性是最低的。
