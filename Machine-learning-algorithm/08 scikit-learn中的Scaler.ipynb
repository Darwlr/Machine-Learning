## Scikit-learn中的Scaler

```python
import numpy as np
from sklearn import datasets
```

```python
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
X[:10, :]
```

```python
# train_test_split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=666)
```

### 1. 首先对 X_train 进行归一化处理

scikit-learn 中的 StandardScaler

```python
from sklearn.preprocessing import StandardScaler
```

```python
standardScaler = StandardScaler()
```

```python
standardScaler.fit(X_train)
```

```python
standardScaler.mean_ # 均值，4列的均值
```

```python
standardScaler.scale_ # 标准差
```

```python
standardScaler.transform(X_train) #将X_train进行归一化处理，这种方式下X_train本身没有改变
```

```python
X_train = standardScaler.transform(X_train)
```

```python
X_train
```

```python
X_test_standard = standardScaler.transform(X_test) # 将测试数据进行归一化
X_test_standard
```

### 2. 将归一化后的数据进行 kNN 分类

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_clf = KNeighborsClassifier(n_neighbors=3)
knn_clf.fit(X_train, y_train)
knn_clf.score(X_test_standard, y_test) #传入的测试数据集必须是归一化后的测试数据集。
```

```python
knn_clf.score(X_test, y_test) # 若没有进行归一化处理
```
