## matplotlib 基础

```python
import matplotlib as mpl # 绘制图形
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100) #[0,10]区间找出100个等分点
```

```python
x
```

```python
y = np.sin(x)
y
```

```python
plt.plot(x, y) # x:横轴；y：纵轴，plt.show()
```

```python
cosy = np.cos(x)
```

```python
siny = y.copy()
```

```python
plt.plot(x, siny) # 在同一坐标系下绘制两条曲线
plt.plot(x, cosy)
```

```python
plt.plot(x, siny) # 在同一坐标系下绘制两条曲线
plt.plot(x, cosy, color = "red", linestyle="--") # 设置颜色
plt.xlim(-5, 15) # 设置x轴范围
plt.ylim(0, 1.5)
```

```python
plt.plot(x, siny) # 在同一坐标系下绘制两条曲线
plt.plot(x, cosy, color = "red", linestyle="--") # 设置颜色
plt.axis([-1, 11, -2, 2]) # x范围[0,10],y范围[-1,1]
```

```python
plt.plot(x, siny) # 在同一坐标系下绘制两条曲线
plt.plot(x, cosy, color = "red", linestyle="--") # 设置颜色
plt.axis([-1, 11, -2, 2]) # x范围[0,10],y范围[-1,1]
plt.xlabel("x axis")
plt.ylabel("y value")
```

```python
plt.plot(x, siny, label= "sin(x)") # 在同一坐标系下绘制两条曲线
plt.plot(x, cosy, color = "red", linestyle="--", label = "cos(x)") # 设置颜色
plt.axis([-1, 11, -2, 2]) # x范围[0,10],y范围[-1,1]
plt.xlabel("x axis")
plt.ylabel("y value")
plt.legend() # 对应 label="sin(x)" label="cos(x)"
```

```python
plt.plot(x, siny, label= "sin(x)") # 在同一坐标系下绘制两条曲线
plt.plot(x, cosy, color = "red", linestyle="--", label = "cos(x)") # 设置颜色
plt.axis([-1, 11, -2, 2]) # x范围[0,10],y范围[-1,1]
plt.xlabel("x axis")
plt.ylabel("y value")
plt.legend() # 对应 label="sin(x)" label="cos(x)"
plt.title("Welcome to the ML world!")
```
